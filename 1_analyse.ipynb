## 4 - Quelle est la différence entre la moyenne mobile et les agrégations précédentes?

La moyenne mobile et les agrégations sont deux techniques couramment utilisées en analyse de séries temporelles, mais elles ont des objectifs différents.

1. **Agrégation** : L'agrégation est le processus de combinaison de plusieurs mesures en une seule mesure.  
 Par exemple, si vous avez des données de consommation d'électricité par heure, vous pouvez les agréger pour obtenir la consommation d'électricité par jour, par mois ou par année.   
   L'agrégation est généralement utilisée pour réduire la granularité des données et faciliter l'analyse à un niveau plus élevé.

2. **Moyenne mobile** : La moyenne mobile, en revanche, est une technique utilisée pour lisser une série temporelle et aider à identifier la tendance sous-jacente en éliminant les fluctuations à court terme.   
Par exemple, une moyenne mobile sur 7 jours calcule la moyenne des 7 jours précédents pour chaque point dans le temps, ce qui a pour effet de lisser les fluctuations quotidiennes et de mettre en évidence les tendances à plus long terme.  

En résumé, l'agrégation est utilisée pour réduire la granularité des données, tandis que la moyenne mobile est utilisée pour lisser les fluctuations à court terme et mettre en évidence les tendances à long terme.

## 6 - Qu’est ce que la stationnarité? 

La stationnarité est une propriété importante des séries temporelles. Une série temporelle est dite stationnaire si ses propriétés statistiques ne changent pas avec le temps.   
En d'autres termes, elle reste constante dans le temps.

Plus précisément, une série temporelle est stationnaire si :

1. La moyenne de la série ne change pas avec le temps. C'est-à-dire que la moyenne est constante sur toute la série.
2. La variance de la série est constante dans le temps. Cela signifie que l'écart moyen des valeurs de la série par rapport à la moyenne reste constant.
3. La covariance entre deux périodes de temps reste constante au fil du temps.

La stationnarité est importante car de nombreux modèles de séries temporelles, comme le modèle ARIMA, supposent que la série est stationnaire.   
  Si une série n'est pas stationnaire, nous devons souvent la transformer pour la rendre stationnaire avant de pouvoir l'utiliser pour modéliser ou prévoir.

### 8 - Réalisez un test de Dickey Fuller, quelle est l’hypothèse nulle et l’hypothèse alternative, quelle est la statistique calculée et quelle distribution théorique suit elle d’après l’hypothèse nulle.   
### Quelle est la probabilité que la statistique sous l’hypothèse H0 prenne un valeur aussi extrême voir plus extrême? 

```python
from statsmodels.tsa.stattools import adfuller

# Rassembler les données par jour et calculer la moyenne
y_daily = df['consommation_brute_electricite_rte'].resample('D').mean()

# Effectuer le test de Dickey-Fuller augmenté
result = adfuller(y_daily.dropna())

# Afficher les résultats
print('Statistique ADF : %f' % result[0])
print('p-value : %f' % result[1])
print('Valeurs Critiques :')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
# Dans ce test :

# L’hypothèse nulle (H0) est que la série temporelle a une racine unitaire, c’est-à-dire qu’elle est non stationnaire.
# L’hypothèse alternative (H1) est que la série temporelle n’a pas de racine unitaire, c’est-à-dire qu’elle est stationnaire.
# La statistique calculée est la statistique ADF (Augmented Dickey-Fuller). Si cette statistique est inférieure à la valeur critique, nous rejetons l’hypothèse nulle.
# La distribution théorique sous l’hypothèse nulle est une distribution normale.
# La p-value représente la probabilité que la statistique sous l’hypothèse H0 prenne une valeur aussi extrême ou plus extrême. Si la p-value est inférieure à un seuil (généralement 0,05), alors nous rejetons l’hypothèse nulle.
```

```python
import seaborn as sns

# Rassembler les données par jour et calculer la moyenne
y_daily = df['consommation_brute_electricite_rte'].resample('D').mean()

# Tracer le graphique de la série temporelle journalière
sns.lineplot(data=y_daily)
plt.title('Série temporelle moyennisée par jour')
plt.show()
```

Les résultats du test de Dickey-Fuller augmenté indiquent que la série temporelle moyennisée par jour est stationnaire. Voici pourquoi :

- La statistique ADF est de -4.236426, ce qui est inférieur aux valeurs critiques à 1%, 5% et 10%. Cela signifie que nous pouvons rejeter l'hypothèse nulle avec un niveau de confiance de plus de 99%.
- La p-value est de 0.000571, ce qui est inférieur au seuil généralement accepté de 0.05. Cela signifie qu'il y a une faible probabilité d'observer une telle statistique si l'hypothèse nulle était vraie.

Par conséquent, nous rejetons l'hypothèse nulle et concluons que la série temporelle est stationnaire. 

## 9- Même consigne pour le test de KPSS


```python
from statsmodels.tsa.stattools import kpss

# Rassembler les données par jour et calculer la moyenne
y_daily = df['consommation_brute_electricite_rte'].resample('D').mean()

# Effectuer le test KPSS
result = kpss(y_daily.dropna())

# Afficher les résultats
print('Statistique KPSS : %f' % result[0])
print('p-value : %f' % result[1])
print('Valeurs Critiques :')
for key, value in result[3].items():
    print('\t%s: %.3f' % (key, value))
```

Les résultats du test KPSS indiquent que la série temporelle moyennisée par jour n'est pas stationnaire. Voici pourquoi :

- La statistique KPSS est de 0.493417, ce qui est supérieure aux valeurs critiques à 10%, 5% et 2.5%, mais inférieure à la valeur critique à 1%.   
Cela signifie que nous pouvons rejeter l'hypothèse nulle avec un niveau de confiance de plus de 95% mais moins de 99%.
- La p-value est de 0.043149, ce qui est inférieure au seuil généralement accepté de 0.05.   
Cela signifie qu'il y a une faible probabilité d'observer une telle statistique si l'hypothèse nulle était vraie.

Par conséquent, nous rejetons l'hypothèse nulle et concluons que la série temporelle n'est pas stationnaire. 

```python
# Tracer le graphique de la série temporelle journalière
sns.lineplot(data=y_daily)
plt.title('Série temporelle moyennisée par jour')
plt.show()
```

### ___________________________________________________________________


Une série temporelle est une suite d'observations d'un même phénomène, ordonnées dans le temps. Les séries temporelles peuvent être décomposées en plusieurs composantes : la tendance, la saisonnalité et la composante résiduelle.

Il existe deux types de modèles pour décomposer une série temporelle : le modèle additif et le modèle multiplicatif.

1. **Modèle additif** : Dans une série temporelle additive, les composants s'additionnent pour former la série. C'est-à-dire que la série temporelle est représentée comme une addition des trois composantes :
    - Série temporelle = Tendance + Saisonnalité + Aléatoire
    - Ce modèle est généralement utilisé lorsque la magnitude de la saisonnalité ne change pas avec le temps.  

    

2. **Modèle multiplicatif** : Dans une série temporelle multiplicative, les composants se multiplient pour former la série. C'est-à-dire que la série temporelle est représentée comme une multiplication des trois composantes :
    - Série temporelle = Tendance * Saisonnalité * Aléatoire
    - Ce modèle est généralement utilisé lorsque la magnitude de la saisonnalité augmente avec le temps.

Par exemple, si vous observez les ventes d'un magasin au fil du temps, vous pouvez remarquer qu'il y a une tendance à la hausse (peut-être due à l'expansion du magasin), une saisonnalité (peut-être due à des périodes de soldes ou des vacances) et une composante résiduelle (peut-être due à des événements imprévus comme une panne de courant). 
- Si l'effet de la saisonnalité reste constant au fil du temps (par exemple, les ventes augmentent toujours de 50 unités pendant les soldes), alors un modèle **additif** pourrait être approprié. 
- Si l'effet de la saisonnalité change avec le temps (par exemple, les ventes doublent pendant les soldes), alors un modèle **multiplicatif** pourrait être plus approprié.



```python
from dateutil.parser import parse
from statsmodels.tsa.seasonal import seasonal_decompose

# Appliquez la décomposition saisonnière avec l'option "extrapolate_trend" définie sur "freq"
result_add = seasonal_decompose(df["consommation_brute_electricite_rte"].resample('M').mean(), model='additive', extrapolate_trend='freq')

# Tracer les composantes de la décomposition
plt.rcParams.update({'figure.figsize': (15, 15)})
result_add.plot().suptitle('Additive Decompose', fontsize=10)
plt.show()
```
